```python
from math import sqrt
def pol(x,y):
    r = np.sqrt(x**2 +y**2)
    phi = np.arctan2(y,x)
    return(r,phi)
x = np.random.randint(100,size=(10,2))
y = np.random.randint(100,size=(10,2))
print(pol(x,y))
```
